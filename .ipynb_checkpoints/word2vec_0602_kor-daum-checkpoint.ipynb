```python
def load_data(path):
    all_daum_txt = open(path, 'r')
    i = 0
    news_id = []
    comment_id = []
    comments = []

    while True:
        line = all_daum_txt.readline()
        if not line: break
        n_id = line.split("|$|")[0] # news_id
        news_id.append(n_id)
        c_id = line.split("|$|")[1] # comment_id
        comment_id.append(c_id)
        temp = line.split("|$|")[4] # comment
        comments.append(temp)
        i += 1

    print (i)
    print ("@@@@@@")
#     i = 0
#     print ("comments: ")
#     for a in comments:
#         print (a)
#         i += 1
#         if i > 10: break

#     # 비어 있는 댓글 제거
#     i = 0
#     for comment in comments:
#         if comment == "":
#             comments.pop(i)
#             print (i)
#         i+=1
        
    daum_data = pd.DataFrame([news_id, comment_id, comments])
    daum_data = daum_data.T
    daum_data.columns = ["news_id", "comment_id", "comment"]
    return daum_data
```

```python
def tokenizer_comment(comment):
    from konlpy.tag import Twitter
    twitter = Twitter()
    i = 0
    malist = twitter.pos(comment)
    r = []
    
    tag = ["Noun", "Adjective"]
    stopwords = ["것", "이", "안", "더", "왜", "때", "좀", "뭐", "거", "저", "뿐", "머"]
    
    try:
        for word in malist:
            # 어미/조사/구두점/ㅋㅋ^^ㅎㅎ/음표살림/Alphabet/부사는 대상에서 제외            
            if word[1] in tag:
                if not word[0] in r:
                    if not word[0] in stopwords:
                        # 숫자, 특수문자 제거.
                        r.append(word[0])
        return r

    except Exception as e:
        print (e)
```

```python
def comment_count(token_data):
    unique_comment_tokenized = [list(i) for i in set(tuple(i) for i in token_data)]
    word_dic = {}
    
    # word count
    for words in unique_comment_tokenized:
        for word in words:
            if not(word in word_dic):
                word_dic[word] = 0
            word_dic[word] += 1

    keys = sorted(word_dic.items(), key = lambda x:x[1], reverse = True)
    for word, count in keys[:500]:
        print("{0}({1}) ". format(word, count), end = "")
    
    # [] 없애주는 코드
    from itertools import chain
    words = set(chain(*unique_comment_tokenized))
        
    n_vocab = len(words)
    print ("")
    print ("Total Vocab: ", n_vocab)
    print ("")
    
    return keys, n_vocab
```

```python
token_data = []
i = 0
for comment in all_daum_data.comment:
    temp = tokenizer_comment(comment)
    token_data.append(temp)
    i += 1                                                                                                 
    
    if i % 100000 == 0:
        print (i)
```

```python
# 부동산, 정치, 일자리
a = model_w2v.most_similar(positive = ["개혁"], topn=1000)

sim_word = []
for item in a:
    sim_word.append(item[0])

adj_word = []
count = 0

def sil_comment(comment):
    from konlpy.tag import Twitter
    twitter = Twitter()
    i = 0
    malist = twitter.pos(comment)
    return malist

for item in sim_word:
    temp = sil_comment(item)
    if temp[0][1] == 'Adjective':
        adj_word.append(temp[0][0])
        count += 1

print ("count: ", count)
adj_word
```
